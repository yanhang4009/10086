```python
from sklearn.cluster import KMeans
import torch
```

```python
write_logits_path = '../logits.pt'
num_cluster = 10

all_consolidated_logits = torch.load(write_logits_path)
print(len(all_consolidated_logits))
for idx, each_tensor in enumerate(all_consolidated_logits):
        all_consolidated_logits[idx] = each_tensor.numpy()

model = KMeans(n_clusters=num_cluster,            # 聚类中心的数量
                init='k-means++',        # 初始化中心点的方法
                n_init=10,               # 不同质心种子的运行次数
                max_iter=300,            # 最大迭代次数
                tol=1e-4,                # 收敛的容忍度
                random_state=42)# 随机数种子，确保可重复性
print(type(all_consolidated_logits))
print(all_consolidated_logits[100].shape)
model.fit(all_consolidated_logits)
y = model.predict(all_consolidated_logits)
idx_ = []
for idx in range(len(y)):
        idx_.append(y[idx])
print(idx_)
# with open('cluster_idx.txt', 'w') as ppp:
#         ppp.write(str(idx_))
# print(len(y))
```

```python
__dic = {}
index = -1
cluster_idx2logit = {}
for kk in idx_:
    index += 1
    if kk in __dic:
        continue
    else:
        __dic[kk] = 1
        cluster_idx2logit[kk] = all_consolidated_logits[index]
        print(all_consolidated_logits[index].shape)
with open("cluster_idx2logit.txt", 'w') as ppp:
    print(type(cluster_idx2logit))
    cluster_idx2logit = sorted(cluster_idx2logit.items(), key = lambda x: x[0])
    target = []
    for i in range(num_cluster):
        target.append(dict(cluster_idx2logit)[i])
    
    # torch.save(target, 'cluster_idx2logit.pt')
    # ppp.write(str(target))
print(len(cluster_idx2logit))
```

```python
cluster_idx2idx = {}
idx = -1
for kk in idx_:
    idx += 1
    if kk in cluster_idx2idx:
        cluster_idx2idx[kk].append(idx)
    else:
        cluster_idx2idx[kk] = [idx]
# with open('cluster_idx2id.txt', 'w') as ppp:
#     ppp.write(str(cluster_idx2idx))
for key,value in cluster_idx2idx.items():
    print(key,':  ', len(value))
```
